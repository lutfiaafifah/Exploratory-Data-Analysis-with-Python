## Load dataset
Load dataset previous stored dataset after preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_car = pd.read_csv('automobile_dataset_preprocessing.csv')

# display records of all columns
pd.set_option('display.max_columns', None)

df_car.head(10)
```

## Remove the first column of dataset which is the index of the previous one

```python
df_car = df_car.drop('Unnamed: 0', axis=1)
df_car.head(10)
```

## Analize the descriptive statistic
The code below shows data distribution of numeric column in dataset

```python
df_car.describe()
```

We can see data distribution of numeric column in dataset that includes count of the data, mean, standar deviation (std), minimum value, maximum value, and quartile (25, 50, and 75 percentile) of each numeric column.

## Detect the outliers in each column

```python
for i in range(len(df_car.columns)):
    if (df_car[df_car.columns[i]].dtypes in ['int64','float64']):
        print('\n',df_car.columns[i],':',df_car[df_car.columns[i]].dtypes)
        Q1=df_car[df_car.columns[i]].quantile(0.25)
        print('Q1',Q1)
        Q3=df_car[df_car.columns[i]].quantile(0.75)
        print('Q3',Q3)
        IQR=Q3-Q1
        print('IQR',IQR)
        min=df_car[df_car.columns[i]].min()
        max=df_car[df_car.columns[i]].max()
        min_IQR=Q1-1.5*IQR
        max_IQR=Q3+1.5*IQR
        print('Lower', min_IQR)
        print('Upper', max_IQR)
        if (min<min_IQR):
            print('Low outlier is found < ',min_IQR,' - ', min)
        if (max>max_IQR):
            print('High outlier is found > ',max_IQR,' - ', max)
```
