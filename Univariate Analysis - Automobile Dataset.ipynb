## Load dataset
Load dataset previous stored dataset after preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_car = pd.read_csv('automobile_dataset_preprocessing.csv')

# display records of all columns
pd.set_option('display.max_columns', None)

df_car.head(10)
```

## Remove the first column of dataset which is the index of the previous one

```python
df_car = df_car.drop('Unnamed: 0', axis=1)
df_car.head(10)
```

## Common Plotting Parameters

```python
plot_color = '#4a91f2'

plt.subplots_adjust(
    left    =  0.1, 
    bottom  =  0.4, 
    right   =  0.2, 
    top     =  0.5, 
    wspace  =  0.2, 
    hspace  =  0.2
)
sns.set_style("whitegrid") #set seaborn style template
```

## symboling Distribution

```python
fig, ax = plt.subplots(figsize=(12,5), ncols=2, nrows=1)

symboling_hist = sns.countplot(df_car["symboling"], color=plot_color, ax=ax[0])
symboling_hist.set_ylabel('Frequency',fontsize=14 )
symboling_hist.set_xlabel('symboling',fontsize=14)
symboling_hist.set_title('symboling - Distribution\n', fontsize=16)
symboling_hist.set_xticklabels(symboling_hist.get_xticklabels(), fontsize=14)

sym_hist = sns.distplot(df_car['symboling'], hist=True, color=plot_color, ax=ax[1])
sym_hist.set_xlabel('\nsymboling',fontsize=14)
sym_hist.set_title('symboling - Distribution\n', fontsize=16)

fig.savefig('symboling_distribution.jpg', dpi=fig.dpi, bbox_inches='tight')
```

```python
df_car['symboling'].value_counts()
```

## make Distribution

```python
fig, ax = plt.subplots(figsize=(12,7), ncols=1, nrows=1)

make_hist = sns.countplot(df_car["make"], color=plot_color)
make_hist.set_ylabel('Count',fontsize=16)
make_hist.set_xlabel('')
make_hist.set_title('make - Distribution\n', fontsize=18)
ax.patch.set_alpha(0.5)

make_hist.set_xticklabels(make_hist.get_xticklabels(), rotation=90, fontsize=14)

fig.savefig('make_distribution.jpg', dpi=fig.dpi, bbox_inches='tight')
```

```python
df_car['make'].value_counts()
```
